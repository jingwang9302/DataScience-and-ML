```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('X axis Title')
plt.ylabel('Y axis Title')
plt.title('String Title')
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1) 
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
# Create Figuare(empty canvas)
fig = plt.figure()
# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, botto,, width,height
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes1.plot(x,y)
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # inset axes
axes2.plot(y,x,'r')
```

```python
# empty canvas of 1  by 2 subplots
fig, axes = plt.subplots(1,2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout
```

## Figure Size and DPI

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1,figsize=(5,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## save image

```python
fig.savefig('my_pic.png',dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X2')
ax.plot(x,x**3,label = 'X3')
ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'orange',lw = 3, ls = ':',marker  ='o',markersize = 10,markerfacecolor = 'yellow') #lw:linewidth, ls: linestyle
```
