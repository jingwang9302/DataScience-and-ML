```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
#sns.despine()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=1.2)
sns.countplot(x='sex',data=tips)
```
